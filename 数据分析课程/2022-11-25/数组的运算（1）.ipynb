### 算术运算符 与 api 都可以使用


```python
# 绝对值 
x = np.array([-2,-1,0,1,2])
abs(x)    # 慢 慢 慢 ！！！！！ 在处理数据时尽量使用 numpy 原生
```

```python
np.absolute(x)
```

```python
# 三角函数
theta = np.linspace(0,np.pi,3)
theta
```

```python
np.sin(theta)
```

```python
np.cos(theta)
```

```python
np.tan(theta)
```

```python
x = [-1,0,1]
# 逆三角函数
np.arcsin(x)
np.arccos(x)
np.arctan(x)
```
